```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import os
import random
```

```python
Data = np.zeros([52, 8])


file_dir = './NEW DATA/DOE' # csv위치 폴더
file_list = os.listdir(file_dir)

for i in range(len(file_list)):
    tmp = pd.read_csv(file_dir + '/' + file_list[i])                           # 파일 불러오기
    Data[i, :4]   = tmp.iloc[0, 1:]                                            # 초기값(모니터링 변수) 저장
    Data[i, 4:-2] = [float(file_list[i][3:6]), float(file_list[i][8:10])]      # P, I값 저장
    Data[i, -2] = tmp.shape[0]                                                 # 전체 공정시간 저장 (아마 사용 X)
    
    idx = np.where(tmp.iloc[:, 0] == 'eSTERILIZATION')[0][0]                   # 멸균 직전 시간까지 index 가져오기
    Data[i, -1] = tmp.iloc[:idx, :].shape[0]                                   # 조정단계까지 시간 저장

file_dir = './NEW DATA/OTHER' # csv위치 폴더
file_list = os.listdir(file_dir)
    
for i in range(len(file_list)):
    tmp = pd.read_csv(file_dir + '/' + file_list[i])                           # 파일 불러오기
    Data[i+45, :4]   = tmp.iloc[0, 1:]                                         # 초기값(모니터링 변수) 저장
    Data[i+45, 4:-2] = [float(file_list[i][3:6]), float(file_list[i][8:10])]   # P, I값 저장
    Data[i+45, -2] = tmp.shape[0]                                              # 전체 공정시간 저장 (아마 사용 X)
    
    idx = np.where(tmp.iloc[:, 0] == 'eSTERILIZATION')[0][0]                   # 멸균 직전 시간까지 index 가져오기
    Data[i+45, -1] = tmp.iloc[:idx, :].shape[0]                                # 조정단계까지 시간 저장
    
Data = pd.DataFrame(Data, columns = ['Plate Temp', 'Steam Temp', 'Vessel Pressure', 'Vessel Temp', 'P', 'I', 'Total Time', 'ADJ Time'])
Data
```

```python
def PI_Value():
    random_temperature = []
    x1 = round(random.uniform(0.1, 4), 1)
    x2 = round(random.uniform(10, 90))
    random_temperature.append(x1)
    random_temperature.append(x2)
    return random_temperature
```

```python
# random.seed(0)

num_points = 2                      #설정 제어값 개수
random_num = 100                    # 몇개의 랜덤한 데이터에 대해서
random_PI = np.zeros((random_num,4+num_points))
random_idx = random.randrange(0, Data.shape[0])

for i in range (random_num):
    random_temp = PI_Value()
    random_PI[i,:4] = Data.iloc[random_idx,:4]  # 순서: 'Plate Temp', 'Steam Temp', 'Vessel Pressure', 'Vessel Temp'
    random_PI[i,4:] = random_temp
    
pd.DataFrame(random_PI).shape
```

```python
Optimization_list = []
iteration = 500

New_Dataset = random_PI

for i in range(iteration):

    random_thickness = pd.DataFrame(LoadModel.predict(New_Dataset, verbose=0))
    random_thickness_rank = random_thickness.sort_values([0],ascending = True)
    Parent1 = New_Dataset[random_thickness_rank.index[0],:]
    Parent2 = New_Dataset[random_thickness_rank.index[1],:]
    
    try:
        if (Parent1 != Old_Parent).any():
            Optimization_list.append(i)
            Optimization_list.append(Parent1)
            print('\n-------------------------- iteration : %d --------------------------\n'%i)
            print('Parent1 :',Parent1)
            print('Parent2 :',Parent2)
            print('%d iteration 최적 P, I : '%(i),New_Dataset[random_thickness_rank.index[0],4:])
            print('최소 시간 :', random_thickness_rank.iloc[0, 0])
            
            
    except NameError:
        Optimization_list.append(i)
        Optimization_list.append(Parent1)
        print('\n-------------------------- iteration : %d --------------------------\n'%i)
        print('Parent1 :',Parent1)
        print('Parent2 :',Parent2)
        print('%d iteration 최적 P, I : '%(i),New_Dataset[random_thickness_rank.index[0],4:])
        print('최소 시간 :', random_thickness_rank.iloc[0, 0])

        
    Old_Parent = Parent1
    New_Dataset = np.zeros((14,random_PI.shape[1]))

    Child1 = np.concatenate([Parent1[:4], [Parent1[4],Parent2[5]]])
    Child2 = np.concatenate([Parent1[:4], [Parent2[4],Parent1[5]]])


    Mutant1 = np.concatenate([Parent1[:4], [Parent1[4], round(random.uniform(10, 90))]])
    Mutant2 = np.concatenate([Parent1[:4], [Parent1[4], round(random.uniform(10, 90))]])
    Mutant3 = np.concatenate([Parent1[:4], [Parent2[4], round(random.uniform(10, 90))]])
    Mutant4 = np.concatenate([Parent1[:4], [Parent2[4], round(random.uniform(10, 90))]])
    Mutant5 = np.concatenate([Parent1[:4], [round(random.uniform(0.1, 4), 1), Parent1[5]]])
    Mutant6 = np.concatenate([Parent1[:4], [round(random.uniform(0.1, 4), 1), Parent1[5]]])
    Mutant7 = np.concatenate([Parent1[:4], [round(random.uniform(0.1, 4), 1), Parent2[5]]])
    Mutant8 = np.concatenate([Parent1[:4], [round(random.uniform(0.1, 4), 1), Parent2[5]]])
    Mutant9 = np.concatenate([Parent1[:4], PI_Value()])
    Mutant10= np.concatenate([Parent1[:4], PI_Value()])

    New_Dataset[0,:] = Parent1
    New_Dataset[1,:] = Parent2
    New_Dataset[2,:] = Child1
    New_Dataset[3,:] = Child2
    New_Dataset[4,:] = Mutant1
    New_Dataset[5,:] = Mutant2
    New_Dataset[6,:] = Mutant3
    New_Dataset[7,:] = Mutant4
    New_Dataset[8,:] = Mutant5
    New_Dataset[9,:] = Mutant6
    New_Dataset[10,:] = Mutant7
    New_Dataset[11,:] = Mutant8
    New_Dataset[12,:] = Mutant9
    New_Dataset[13,:] = Mutant10
```
